```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['rapido_cleaned_data.csv']))
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(6, 4))
sns.histplot(df["trip_duration"], bins=30, kde=True, color="blue")
plt.title("Trip Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df["fare_amount"], bins=30, kde=True, color="green")
plt.title("Fare Amount Distribution")
plt.xlabel("Fare (₹)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="start_hour", data=df, palette="coolwarm")
plt.title("Peak Booking Hours")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Trips")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="day_of_week", data=df,
              order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
              palette="magma")
plt.title("Bookings Per Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Trips")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x="pickup_location", y="fare_amount", data=df, palette="coolwarm")
plt.title("Fare Amount by Pickup Location")
plt.xlabel("Pickup Location")
plt.ylabel("Fare Amount (₹)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="payment_mode", data=df, palette="pastel")
plt.title("Payment Mode Distribution")
plt.xlabel("Payment Mode")
plt.ylabel("Number of Transactions")
plt.show()
```
